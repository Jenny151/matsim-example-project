## Strategy 1: Place next to certain locations

```python
import pandas as pd
```

```python
import requests
from requests.structures import CaseInsensitiveDict

url = "https://api.geoapify.com/v2/places?categories=commercial.supermarket&filter=rect%3A10.716463143326969%2C48.755151258420966%2C10.835314015356737%2C48.680903341613316&apiKey=6d0183ed655a4c4bbc7bd1021ce5540e"

headers = CaseInsensitiveDict()
headers["Accept"] = "application/json"

resp = requests.get(url, headers=headers)

print(resp.status_code)
print(resp.text)
```

```python
df_supermarkets = pd.DataFrame(columns=['name', 'long', 'lat', 'city'])

supermarket_names = []
supermarket_longitude = []
supermarket_latitude = []
supermarket_city = []
for i in range(15):
  supermarket_names.append(resp.json()['features'][i]['properties']['name'])
  supermarket_longitude.append(resp.json()['features'][i]['properties']['lon'])
  supermarket_latitude.append(resp.json()['features'][i]['properties']['lat'])
  supermarket_city.append(resp.json()['features'][i]['properties']['city'])

df_supermarkets['name'] = supermarket_names
df_supermarkets['long'] = supermarket_longitude
df_supermarkets['lat'] = supermarket_latitude
df_supermarkets['city'] = supermarket_city
df_supermarkets
```
